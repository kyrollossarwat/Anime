### 3] Data Preprocessing

```python
#Check Null values and hundle it
df1.isna().sum()
```

```python
#Check Duplicate and hundle it
df1.duplicated().sum()
```

```python
#Drop unnecessary columns
df1.drop(['Description', 'Synonyms','Japanese','Producers', 'Licensors','Demographic'],axis=1 , inplace=True)
```

```python
df1.dropna(subset=['English'],inplace=True)
```

```python
df1.isna().sum()
```

```python
df1.drop("Broadcast",axis = 1 , inplace=True)
```

```python
df1['Premiered']
```

```python
#add column
df1["Year"] = df1["Premiered"].str[-4:]
```

### 4] Data Analysis & Visualization

```python
  # 1) Top 10 Anime by Popularity
Top10_anime=df1.groupby("English")[["Popularity"]].sum().sort_values("Popularity",ascending=False).reset_index().head(10)
Top10_anime
```

```python
#bar plot for Top 10 Anime with x-axis: English and Y-axis: Popularity
plt.figure(figsize=(8,4))
plt.bar(Top10_anime["English"],Top10_anime["Popularity"],color="blue")
plt.title("Top 10 Anime by popularity")
plt.xlabel("Anime")
plt.ylabel("Popularity")
plt.xticks(rotation=90)
plt.show()
```

```python
# 2)Score distribution
plt.figure(figsize=(8,4))
sns.histplot(df1["Score"] , color="blue")
plt.title("Score Distribution")
plt.show()
```

```python
######## 3) Yearwise Members ############
# TOP 7 years by total numbers
Top_7_years = df1.groupby("Year")[["Members"]].sum().sort_values("Members",ascending=False).reset_index().head(7)
Top_7_years
```

```python
plt.figure(figsize=(8,4))
plt.bar(Top_7_years["Year"],Top_7_years["Members"],color="blue")
plt.title("Top 7 Years by Total Members")
plt.xlabel("Year")
plt.ylabel("Members")
plt.show()
```

```python
# 4)Type
#Output::
Type
TV       569
Movie    235
OVA       84
```

```python
Type_distribution=df1.groupby("Type").size().reset_index().rename(columns={0:"Count"}).sort_values("Count",ascending=False).head(3)
Type_distribution
```

```python
#Pie Plot
plt.figure(figsize=(8,4))
plt.pie(Type_distribution["Count"],labels=Type_distribution["Type"])
plt.title("Type DIstribution")
plt.show()
```

```python
# 5) Genres
#Output::
Genres
ActionAction, AdventureAdventure, FantasyFantasy    43
ComedyComedy, RomanceRomance                        28
ActionAction, FantasyFantasy                        23
ActionAction, Sci-FiSci-Fi                          22
DramaDrama, RomanceRomance                          21
```

```python
Genre_distribution=df1.groupby("Genres").size().reset_index().rename(columns={0:"Count"}).sort_values("Count",ascending=False).head(5)
Genre_distribution
```

```python
#bar Plot
plt.figure(figsize=(8,4))
plt.bar(Genre_distribution["Genres"],Genre_distribution["Count"])
plt.title("Top 5 Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# 6) Top Studios
Top_Studios=df1.groupby("Studios").size().reset_index().rename(columns={0:"Count"}).sort_values("Count",ascending=False).head(5)
Top_Studios
```

```python
#barplot
plt.figure(figsize=(8,4))
sns.barplot(x=Top_Studios["Studios"], y=Top_Studios["Count"], color="blue")
plt.title("Top 5 Studios")
```

```python
Top_Sources=df1.groupby("Source").size().reset_index().rename(columns={0:"Count"}).sort_values("Count",ascending=False).head(4)
Top_Sources
```
